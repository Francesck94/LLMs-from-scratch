# Exercise 4.2 Initializing larger GPT models <br>
We initialized a 124-million-parameter GPT model, which is known as “GPT-2 small.”
Without making any code modifications besides updating the configuration file, use
the GPTModel class to implement GPT-2 medium (using 1,024-dimensional embeddings,
24 transformer blocks, 16 multi-head attention heads), GPT-2 large (1,280-
dimensional embeddings, 36 transformer blocks, 20 multi-head attention heads),
and GPT-2 XL (1,600-dimensional embeddings, 48 transformer blocks, 25 multi-head
attention heads). As a bonus, calculate the total number of parameters in each GPT
model.

```python
# main.py
import sys
sys.path.insert(0, "../01_main-chapter-code")  # add Folder_2 path to search list

from gpt import GPTModel
```

```python
GPT_CONFIG_SMALL = {
    "vocab_size": 50257,    # Vocabulary size
    "context_length": 1024, # Context length
    "emb_dim": 768,         # Embedding dimension
    "n_heads": 12,          # Number of attention heads
    "n_layers": 12,         # Number of layers
    "drop_rate": 0.1,       # Dropout rate
    "qkv_bias": False       # Query-Key-Value bias
}
```

```python
GPT_CONFIG_MEDIUM = {
    "vocab_size": 50257,    # Vocabulary size
    "context_length": 1024, # Context length
    "emb_dim": 1024,         # Embedding dimension
    "n_heads": 16,          # Number of attention heads
    "n_layers": 24,         # Number of layers
    "drop_rate": 0.1,       # Dropout rate
    "qkv_bias": False       # Query-Key-Value bias
}
```

```python
GPT_CONFIG_LARGE = {
    "vocab_size": 50257,    # Vocabulary size
    "context_length": 1024, # Context length
    "emb_dim": 1280,         # Embedding dimension
    "n_heads": 20,          # Number of attention heads
    "n_layers": 36,         # Number of layers
    "drop_rate": 0.1,       # Dropout rate
    "qkv_bias": False       # Query-Key-Value bias
}
```

```python
GPT_CONFIG_X_LARGE = {
    "vocab_size": 50257,    # Vocabulary size
    "context_length": 1024, # Context length
    "emb_dim": 1600,         # Embedding dimension
    "n_heads": 25,          # Number of attention heads
    "n_layers": 48,         # Number of layers
    "drop_rate": 0.1,       # Dropout rate
    "qkv_bias": False       # Query-Key-Value bias
}
```

```python
def get_num_params(model, model_name):
    total_params = sum(p.numel() for p in model.parameters())
    print(f"{model_name} - Total number of parameters: {total_params:,}")
```

## instantiation of models

```python
gpt_medium = GPTModel(GPT_CONFIG_MEDIUM)
```

```python
get_num_params(gpt_medium, 'gpt_medium')
```

```python
gpt_large = GPTModel(GPT_CONFIG_LARGE)
get_num_params(gpt_large, 'gpt_large')
```

```python
gpt_x_large = GPTModel(GPT_CONFIG_X_LARGE)
get_num_params(gpt_x_large, 'gpt_x_large')
```
